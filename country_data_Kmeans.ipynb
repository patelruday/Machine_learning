```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
x=country.iloc[:,1:]
```

```python
x.head()
```

```python
scaler=MinMaxScaler()
```

```python
country_sc=scaler.fit_transform(x)
```

```python
type(country_sc)
```

```python
df_sc=pd.DataFrame(country_sc,columns=x.columns,index=country['country'])
```

```python
df_sc.head()
```

```python
km=KMeans(n_clusters=3,random_state=2024)
```

```python
km.fit(country_sc)
```

```python
y_pred=km.predict(country_sc)
```

```python
y_pred
```

```python
df_sc['cluster']=y_pred
```

```python
df_sc.head()
```

```python
df1=df_sc[df_sc['cluster']==0]
df2=df_sc[df_sc['cluster']==1]
df3=df_sc[df_sc['cluster']==2]
plt.scatter(df1['inflation'],df1['gdpp'],color='green')
plt.scatter(df2['inflation'],df2['gdpp'],color='red')
plt.scatter(df3['inflation'],df3['gdpp'],color='yellow')
plt.scatter(km.cluster_centers_[:,5],km.cluster_centers_[:,8],color='purple',marker='*',label='cetroid')
plt.xlabel('inflation')
plt.ylabel('gdpp')
plt.legend()
```

```python
plt.scatter(df1['child_mort'],df1['gdpp'],color='green')
plt.scatter(df2['child_mort'],df2['gdpp'],color='red')
plt.scatter(df3['child_mort'],df3['gdpp'],color='yellow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,8],color='purple',marker='*',label='cetroid')
plt.xlabel('child_mort')
plt.ylabel('gdpp')
plt.legend()
```
